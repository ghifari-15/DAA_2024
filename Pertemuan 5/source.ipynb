```python
# Function to find the summation using Divide and Conquer
def sum_array(arr, low, high):
    # Base case: if there's only one element
    if low == high:
        return arr[low]
    
    # Divide the array into two parts
    mid = (low + high) // 2
    
    # Recursively sum the two halves
    left_sum = sum_array(arr, low, mid)
    right_sum = sum_array(arr, mid + 1, high)
    
    # Combine the results of the two halves
    return left_sum + right_sum


arr = [12, 51, 23, 84, 5, 9]

low = 0
high = len(arr) - 1
total_sum = sum_array(arr, low, high)
print(f"Total sum: {total_sum}")
    

if total_sum == sum(arr):  
    print("Total sum matched!") 
```

```python
def minmax(arr, low, high):
    class Pair:
        def __init__(self):
            self.max = 0
            self.min = 0
    result = Pair()
    # If only one element in the array
    if low == high:
        result.max = arr[low]
        result.min = arr[low]
        return result
    # If there are two elements in the array
    if high == low + 1:
        if arr[low] < arr[high]:
            result.min = arr[low]
            result.max = arr[high]
        else:
            result.min = arr[high]
            result.max = arr[low]
        return result
    # If there are more than two elements in the array
    mid = (low + high) // 2
    left = minmax(arr, low, mid)
    right = minmax(arr, mid + 1, high)
    # Compare and get the maximum of both parts
    result.max = max(left.max, right.max)
    # Compare and get the minimum of both parts
    result.min = min(left.min, right.min)
    return result

awokawok = [10, 4, 2, 5, 6, 7, 10, 4, 3]
result = minmax(awokawok, 0, len(awokawok) - 1)
print(f"Max result is: {result.max} and minimum is {result.min}")

```

```python
def insertion_sort(arr):
    # Traverse through 1 to len(arr)
    for i in range(1, len(arr)):
        key = arr[i]
        # Move elements of arr[0..i-1], that are greater than key,
        # to one position ahead of their current position
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

# Example usage
arr = [12, 11, 13, 5, 6]
print("Original array:", arr)
insertion_sort(arr)
print("Sorted array:", arr)
```

```python
def quick_sort(arr, low, high):
    if low < high:
        # Partition the array around the pivot
        pi = partition(arr, low, high)
        
        # Recursively sort elements before and after partition
        quick_sort(arr, low, pi - 1)
        quick_sort(arr, pi + 1, high)

def partition(arr, low, high):
    # Choose the middle element as pivot
    mid = low + (high - low) // 2
    pivot = arr[mid]

    # Move pivot to the end temporarily
    arr[mid], arr[high] = arr[high], arr[mid]
    
    i = low - 1  # Index of smaller element
    
    for j in range(low, high):
        # If current element is smaller than or equal to pivot
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    
    # Place pivot at its correct position
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    
    return i + 1

# Example usage
arr = [10, 7, 8, 9, 1, 5]
n = len(arr)
print("Original array:", arr)
quick_sort(arr, 0, n - 1)
print("Sorted array:", arr)
```

```python
# Hitung Count Inversion dengan divide dan conquer
def countInversion(arr):
    icount = 0
    if len(arr) <= 1:
        return icount

    mid = len(arr) // 2
    left = arr[:mid]
    right = arr[mid:]

    icount += countInversion(left)
    icount += countInversion(right)
    i = j = k = 0

    # Merge the two halves and count inversions
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            arr[k] = left[i]
            i += 1
        else:
            arr[k] = right[j]
            j += 1
            icount += (mid - i)
        k += 1

    # Copy remaining elements from left and right subarrays if any
    while i < len(left):
        arr[k] = left[i]
        i += 1
        k += 1

    while j < len(right):
        arr[k] = right[j]
        j += 1
        k += 1

    return icount

arr = [1, 20, 6, 4, 5]
countInversion(arr)
print(arr)
```

```python
# Tanpa divide dan conquer
def maxSubSum(arr):
    max_so_far = 0
    max_ending_here = 0
    for i in range(len(arr)):
        max_ending_here += arr[i]
        if max_ending_here > max_so_far:
            max_so_far = max_ending_here
        if max_ending_here < 0:
            max_ending_here = 0
    return max_so_far

# Contoh penggunaan
arr = [-2, -5, 6, -2, -3, 1, 5, -6]
result = maxSubSum(arr)
print(result)
```

```python
def maxCrossingSum(arr, low, mid, high):
    result = 0
    leftSum = float('-infinity')
    for i in range(mid, low-1, -1):
        result += arr[i]
        if result > leftSum:
            leftSum = result
    result = 0
    rightSum = float('-infinity')
    for i in range(mid+1, high+1):
        result += arr[i]
        if result > rightSum:
            rightSum = result
    return leftSum + rightSum

def maxSum(arr, low, high):
    if low == high:
        return arr[low]
    mid = (low + high) // 2
    return max(maxSum(arr, low, mid), maxSum(arr, mid+1, high), maxCrossingSum(arr, low, mid, high))

arr = [-2, 5, 6, -2, -3, 1, 5, 6]

result = maxSum(arr, 0, len(arr)-1)

print(result)
```

```python
def longestCommonPrefix(a):
    size = len(a)

    # if size is 0, return empty string
    if (size == 0):
        return ""

    if (size == 1):
        return a[0]

    # sort the array of strings
    a.sort()

    # find the minimum length from first and last string
    end = min(len(a[0]), len(a[size-1]))

    # find the common prefix between the first and last string
    i = 0
    while (i < end and a[0][i] == a[size-1][i]):
        i += 1

    pre = a[0][0:i]
    return pre

arr=["geeksforgeeks", "geeks", "geek", "geezer"]
result=longestCommonPrefix(arr)
print(result)

arr=["apple", "ape", "april"]
result=longestCommonPrefix(arr)
print(result)
```

```python
def medianOfArray(arr1, arr2, n):
    m1 = -1   # first number
    m2 = -1   # second number

    count = 0
    i = j = 0

    while count < n+1:
        count += 1

        if i == n:
            m1 = m2
            m2 = arr2[0]
            break
        if j == n:
            m1 = m2
            m2 = arr1[0]
            break

        if arr1[i] < arr2[j]:
            m1 = m2
            m2 = arr1[i]
            i += 1
        else:
            m1 = m2
            m2 = arr2[j]
            j += 1

    return (m1 + m2)//2

arr1 = [1, 12, 15, 26, 38]
arr2 = [2, 13, 17, 38, 45]

print(medianOfArray(arr1, arr2, len(arr1)))
```

```python
def Solution(arr):
    n = len(arr)

    # If length of array is even
    if n % 2 == 0:
        z = n // 2
        e = arr[z]
        q = arr[z - 1]
        ans = (e + q) / 2
        return ans

    # If length of array is odd
    else:
        z = n // 2
        ans = arr[z]
        return ans

arr1 = [-5, 3, 6, 12, 15]
arr2 = [-12, -10, -6, -3, 4, 10]

# Concatenating the two arrays
arr3 = arr1 + arr2

# Sorting the resultant array
arr3.sort()

print("Median = ", Solution(arr3))
```

```python
def floorSorted(arr, low, high, x):
  """
  Mencari nilai terbesar dalam array yang lebih kecil atau sama dengan x.

  Args:
    arr: Array yang sudah terurut.
    low: Indeks awal dari subarray yang sedang dicari.
    high: Indeks akhir dari subarray yang sedang dicari.
    x: Nilai yang ingin dicari floor-nya.

  Returns:
    Nilai terbesar dalam array yang lebih kecil atau sama dengan x, atau -1 jika tidak ada.
  """

  # Jika low lebih besar dari high, berarti tidak ada elemen yang memenuhi kriteria
  if low > high:
    return -1

  # Jika elemen pertama lebih besar dari x, maka tidak ada elemen yang memenuhi kriteria
  if arr[low] > x:
    return -1

  # Jika elemen terakhir lebih kecil atau sama dengan x, maka elemen terakhir adalah floor-nya
  if arr[high] <= x:
    return arr[high]

  # Cari nilai tengah
  mid = (low + high) // 2

  # Jika elemen tengah sama dengan x, maka elemen tengah adalah floor-nya
  if arr[mid] == x:
    return arr[mid]

  # Jika elemen tengah lebih besar dari x dan elemen sebelumnya lebih kecil dari x,
  # maka elemen sebelumnya adalah floor-nya
  if mid > 0 and x >= arr[mid-1] and arr[mid] > x:
    return arr[mid-1]

  # Jika elemen tengah lebih kecil dari x dan elemen selanjutnya lebih besar dari x,
  # maka elemen tengah adalah floor-nya
  if mid < high and x < arr[mid+1] and x >= arr[mid]:
    return arr[mid]

  # Jika x lebih besar dari elemen tengah, maka cari di subarray sebelah kanan
  if x > arr[mid]:
    return floorSorted(arr, mid+1, high, x)

  # Jika tidak, cari di subarray sebelah kiri
  else:
    return floorSorted(arr, low, mid-1, x)

# Contoh penggunaan
arr = [1, 2, 8, 10, 12, 14, 19]
x = 5

result = floorSorted(arr, 0, len(arr)-1, x)
print(result)  # Output: 2
```

```python
def closestNumber(arr, low, high, x):
    """
    Mencari nilai terdekat dengan x dalam array yang sudah terurut.

    Args:
        arr: Array yang sudah terurut.
        low: Indeks awal dari subarray yang sedang dicari.
        high: Indeks akhir dari subarray yang sedang dicari.
        x: Nilai yang ingin dicari nilai terdekatnya.

    Returns:
        Nilai terdekat dengan x dalam array.
    """

    # Jika indeks awal lebih besar dari indeks akhir, berarti tidak ada elemen yang memenuhi
    if low > high:
        return -1

    # Jika elemen terakhir lebih kecil atau sama dengan x, maka elemen terakhir adalah yang terdekat
    if arr[high] <= x:
        return arr[high]

    # Jika elemen pertama lebih besar dari x, maka elemen pertama adalah yang terdekat
    if arr[low] > x:
        return arr[low]

    # Cari nilai tengah
    mid = (low + high) // 2

    # Jika elemen tengah sama dengan x, maka sudah ketemu
    if arr[mid] == x:
        return arr[mid]

    # Hitung selisih mutlak antara elemen tengah dengan x
    abs_mid = abs(arr[mid] - x)

    # Jika elemen tengah lebih besar dari x:
    if arr[mid] > x:
        # Hitung selisih mutlak antara elemen sebelum tengah dengan x
        abs_left = abs(arr[mid - 1] - x)

        # Jika selisih elemen sebelum tengah lebih kecil, cari di sebelah kiri
        if abs_left < abs_mid:
            return closestNumber(arr, low, mid - 1, x)

    # Jika elemen tengah lebih kecil dari x:
    else:
        # Hitung selisih mutlak antara elemen setelah tengah dengan x
        abs_right = abs(arr[mid + 1] - x)

        # Jika selisih elemen setelah tengah lebih kecil, cari di sebelah kanan
        if abs_right < abs_mid:
            return closestNumber(arr, mid + 1, high, x)

    # Jika tidak ada yang lebih dekat, maka elemen tengah adalah yang terdekat
    return arr[mid]

# Contoh penggunaan
arr = [2, 5, 6, 7, 8, 8, 9]
x = 9
print(closestNumber(arr, 0, len(arr) - 1, x))
```

```python
def fixedPoint(arr, low, high):
    if low > high:
        return -1

    if arr[high] == high:
        return arr[high]

    if arr[low] == low:
        return arr[low]

    mid = (low + high) // 2

    if arr[mid] == mid:
        return arr[mid]

    if mid > arr[mid]:
        return fixedPoint(arr, mid+1, high)

    else:
        return fixedPoint(arr, low, mid-1)

arr=[9,1,4,5,2]

print(fixedPoint(arr, 0, len(arr)-1))
```
